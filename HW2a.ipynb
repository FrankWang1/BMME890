```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
%matplotlib inline
```

```python
df.Age.plot(kind='hist', bins=100) # histrogram of Age in the training set
```

```python
x=df.Pclass
y=df.Fare 
plt.scatter(x,y)   # scatter plot of Pclass and Fare
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.yticks(np.arange(min(y), max(y)+1,30))
plt.xlabel('Pclass')
plt.ylabel('Fare')    
```

```python
pd.crosstab(df.Pclass,df.Survived) # crosstab Pclass and Survival 
```
